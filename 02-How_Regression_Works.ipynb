# How regression works

Regression is a statistical technique used to analyze the relationship between a dependent variable and one or more independent variables. The goal of regression is to find the best-fitting line or curve that can predict the value of the dependent (target) variable based on the values of the independent variables. Regression can be used for a wide range of applications, such as predicting sales revenue based on advertising spending, estimating the impact of education on income, and modeling the relationship between temperature and crop yields.


The formula for a simple linear regression model with one independent variable can be expressed as: $y = b + w x$

where:
* $y$ is the dependent variable (the variable being predicted)
* $x$ is the independent variable (the variable used to predict $y$)
* $b$ is the y-intercept (the value of $y$ when $x = 0$)
* $w$ is the slope of the line (the change in $y$ for a one-unit change in $x$)

To create a regression model:
* Data is collected on both the dependent and independent variables.
* The data is then used to estimate the values of $b$ and $w$ that minimize the sum of the squared errors between the predicted values and the actual values of the dependent variable. This process is known as least squares regression.
* Once the regression model has been estimated, it can be used to predict the value of the dependent variable for a given value of the independent variable.
* The accuracy of the predictions can be evaluated using various statistical measures such as the coefficient of determination (R-squared) and the root-mean-square error (RMSE).

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()   # setting the default Seaborn theme
```

```python
# input variable (house size in 1000s of square feet)
x_train = np.array([1, 1.5, 2, 2.5, 3, 3.5, 4])

# target (in 1000s of dollars)
y_train = np.array([120., 180., 170., 210., 250., 260., 270.])

print('The number of training examples:', len(x_train)) # the number of training example

i = 1  # change this to any i to see the i-th point
point = "(x^(%d), y^(%d)) = (%d, %d)" % (i, i, x_train[i] , y_train[i])
print(point)
```

## Plot the data
Let's plot the points to have a visualization of data.

```python
plt.scatter(x_train, y_train, marker="x", c="b")  # plot the data points
plt.title("Housing prices")                       # set the title
plt.ylabel("Price in 1000s of $")                 # set the y axis label
plt.xlabel("House size (in 1000s of sqft)")       # set the x axis label
plt.show()
```

##  Create a linear function
Create a function that maps from `x` to `y`.

$ f(x) = b + wx$

```python
# Let's try some value for 'b' and 'w': f(x) = b + wx
b = 100
w = 100

# calculate the value of f(x) for any input data
i = 1
f = b + w * x_train[i]
print('actual =', y_train[i])
print('prediction =', f)
```

```python
plt.figure(figsize=(7,5))
plt.plot(x_train, prediction, c="r", label='Our prediction')          # plot the prediction model
plt.scatter(x_train, y_train, marker="x", c="b", label='Actual data') # plot the data points

plt.ylabel("Price in 1000s of $")            # set the y axis label
plt.xlabel("House size (in 1000s of sqft)")  # set the x axis label
plt.legend()                                 # shoe the legend
plt.show()
```
