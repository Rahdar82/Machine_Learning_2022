```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
sns.set()   # setting the default Seaborn theme
```

Befor getting to the cost function of a regression model, I would like to plot $x^2 + y^2$ in a 3d space. So, we can also see how the contours of this function help us find the minimum of a plot.

```python
# create two NumPy arrays values between -10 and 10, with a step size of 0.5
x = np.arange(-10, 10, 0.5)
y = np.arange(-10, 10, 0.5)
```

`np.meshgrid` is a NumPy function that returns coordinate matrices from coordinate vectors.
Given two 1D arrays, it returns two 2D matrices. The first matrix represents the X-coordinates of the points on a grid and the second matrix represents the Y-coordinates of the points on a grid.

```python
# create a rectangular grid out of two given one-dimensional arrays
X, Y = np.meshgrid(x,y)

print(X.shape, '\n')
print('X =\n', X, '\n')
print('Y =\n', Y)
```

```python
Z = X ** 2 + Y ** 2    # assume that this is the cost function to be minimized
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.rainbow)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.rainbow)
cset = ax.contourf(X, Y, Z, zdir='z', offset=0, cmap=plt.cm.rainbow)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.rainbow)
cset = ax.contour(X, Y, Z, zdir='z', offset=0, cmap=plt.cm.rainbow) # instead of ax.contourf, use ax.contour
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

Now, lets try an example.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  # higher level plotting tools
%matplotlib inline
sns.set()
```

```python
# some simple random (close to linear) data
x = np.arange(1,11)
y = x + 1.5 * np.random.randn(10)
print('y =', y)
```

```python
a0 = 0                    # intercept
a1 = 0.5                  # slope
f = lambda x : a0 + a1*x  # fitted line

plt.plot(x, y, 'o', color='g', label='training data')
plt.plot(x, f(x), label='f(x) = %.1f + %.1f*x' %(a0,a1))
plt.xlim([0,11])

for i in range(len(x)):
    plt.plot([x[i], x[i]], [f(x[i]), y[i]], ls='--', color='r', lw=.7)
plt.legend()
plt.show()
```

```python
# The function J
def cost(a0, a1, x, y):
    m = len(x)   # number of data points
    J = 0
    for i in range(m):
        J += ((a0 + a1*x[i]) - y[i] )**2
    return J

a = np.linspace(.5, 2, 100)
plt.plot(a, cost(a0, a ,x ,y))  # for the fixed a0, draw cost for various a1 between 0.5 and 2
plt.show()
```

```python
fig = plt.figure(figsize=(7,5))
plt.plot(x,y, 'o', color='g', label='training data')
a = np.linspace(0.5, 2 ,5)

for i in range(len(a)):
    plt.plot(x, a0 + a[i]*x, label='a1 = %.2f' %a[i] )

plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.gca(projection='3d')
a0 = np.linspace(-2,2,100)
a1 = np.linspace(0,2.2,100)
aa0, aa1 = np.meshgrid(a0, a1)
surf = ax.plot_surface(aa0, aa1, cost(aa0,aa1,x,y), cmap=plt.cm.rainbow)
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
a0 = np.linspace(-2,2, 100)
a1 = np.linspace(0,2.2, 100)
fig = plt.figure(figsize=(12, 8))
aa0, aa1 = np.meshgrid(a0, a1)
plt.contour(aa0,aa1,cost(aa0,aa1,x,y) , levels=[i for i in np.arange(0,80,3.75)], cmap=plt.cm.rainbow)
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
