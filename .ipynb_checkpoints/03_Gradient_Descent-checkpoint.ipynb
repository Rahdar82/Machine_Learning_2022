# Gradient Descent

In this lab, you learn how optimize $b$ and $w$ in linear regression using gradient descent algorithm.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate some random data
x = 2 * np.random.rand(100,1)
y = 5 + 3 * x + np.random.randn(100,1)

plt.plot(x,y,'b.')
plt.xlabel("$x$", fontsize=14)
plt.ylabel("$y$", rotation=0, fontsize=14)
plt.grid()
plt.show()
```

```python
def compute_gradient(x, y, t0, t1):
    """
    Computes the gradient for linear regression
    
    Args:
        x  (ndarray): shape (m,) variable such as house size
        y  (ndarray): shape (m,) target value
        t0 (scallar):            initial values of the parameter
        t1 (scallar):            initial values of the parameter
        
    Returns:
        dj_dt0 (scallar): the gradient of the cost with respet to the parameter t0
        dj_dt1 (scallar): the gradient of the cost with respet to the parameter t1
    """
    m = len(x)
    dj_dt0 = 0
    dj_dt1 = 0
    
    for i in range (m): # loop over examples
        f_i = t0 + t1 * x[i]            # calculate prediction for sample i
        dj_dt0_i = f_i - y[i]           # calculate partial derivative
        dj_dt1_i = (f_i - y[i]) * x[i]  # calculate partial derivative
        
        dj_dt0 = dj_dt0 + dj_dt0_i      # add to total
        dj_dt1 = dj_dt1 + dj_dt1_i      # add to total
        
    dj_dt0 = 1/m * dj_dt0           # divide by number of examples
    dj_dt1 = 1/m * dj_dt1           # divide by number of examples
       
    return dj_dt0, dj_dt1
```

```python
iteration = 5000
t0 = 0
t1 = 1
alpha = .01  # learning rate
e = 1e-6     # stopping threshold
m = len(x)
cost = np.zeros((iteration,1))

t_history = np.zeros((iteration,2))
t_history[0,0] = t0
t_history[0,1] = t1

cost[0] = 1/2/m * sum(((t0 + t1 * x) - y)**2)

for i in range(1,iteration):
    
    dj_dt0, dj_dt1 = compute_gradient(x, y, t_history[i-1,0], t_history[i-1,1])
        
    t_history[i,0] = t_history[i-1,0] - (alpha * dj_dt0)
    t_history[i,1] = t_history[i-1,1] - (alpha * dj_dt1)
    
    cost[i] = 1/2/m * sum(((t_history[i,0] + t_history[i,1] * x) - y)**2)    
    
    if abs(cost[i] - cost[i-1]) <= e:
        print("number of iterations:", i)
        print("cost:", cost[i])
        print("derivatives:", dj_dt0, dj_dt1)
        t_history = t_history[:i+1,:]
        t0 = t_history[i,0]
        t1 = t_history[i,1]
        break
    
print("theta0: %.4f" % t0, "\ntheta1: %.4f" % t1)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,4), constrained_layout=True)

ax1.plot(np.arange(i),cost[:i])   # plot cost versus iteration
ax1.set_ylabel("cost")
ax1.set_xlabel("Iterations")
ax1.grid()


ax2.plot(np.arange(150,i),cost[150:i])   # plot cost versus iteration that does not include the initial iterations
ax2.set_ylabel("cost")
ax2.set_xlabel("Iterations")
ax2.grid()
```

```python
# calculate cost for different theta0 and theta1 combinations to plot contour
theta0_vals = np.linspace(0,10,100)
theta1_vals = np.linspace(0,10,100)
J_vals = np.zeros((len(theta0_vals), len(theta1_vals)))

for i in range(len(theta0_vals)):
    for j in range(len(theta1_vals)):
        J_vals[i,j] = 1/2/m * sum(((theta0_vals[i] + theta0_vals[j] * x) - y)**2)
```

```python
fig = plt.figure(figsize=(6, 4))
cont = plt.contour(theta0_vals, theta1_vals, J_vals, cmap=plt.cm.rainbow, levels = 25)
fig.colorbar(cont, shrink=0.5, aspect=5)
plt.xlabel("$\u03B8_0$", fontsize = 14)
plt.ylabel("$\u03B8_1$", fontsize = 14)

plt.show()
```

```python
m = x.shape      # find the shape of x
f = np.zeros(m)  # initialize array f
for i in range(len(x)):
    f[i] = t0 + t1 * x[i]

plt.plot(x,y,'.b')                            # plot the points
plt.plot(x, f, c="r", label="Our prediction") # plot the line

plt.xlabel("$x$", fontsize=14)
plt.ylabel("$y$", rotation=0, fontsize=14)
plt.grid()
plt.show()
```
