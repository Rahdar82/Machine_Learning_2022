# Polynomial Regression
Let's find a quadratic relationship for the house price example: $f(x)=b + w_1 x + w_2 x^2$

Implementing polynomial regression with scikit-learn is very similar to linear regression. There‚Äôs only one extra step: you need to transform the array of inputs to include nonlinear terms such as $ùë•¬≤$.

```python
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
```

```python
x = np.array([1000,1500,1200,1800,2500,800,1600,2400,3100,3800,4200,4500])  # input variable (house size)
y = np.array([100,230,180,180,170,120,240,195,220,270,260,380])    # target (in 1000s of dollars)

x = x.reshape(-1,1)  # reshapes the x array into a 2D array with one column

x_ = x**2            # squaring each element in the original x array

# each row in the resulting x array will contain the original house size and its square
x2 = np.concatenate((x, x_),1)

y = y.reshape(-1,1)  # reshapes the y array into a 2D array with one column
```

```python
# Create StandardScaler objects for x and y
x_scaler = StandardScaler()
y_scaler = StandardScaler()

# Fit and transform the input features (x) and the target variable (y)
x_scaled = x_scaler.fit_transform(x2)
y_scaled = y_scaler.fit_transform(y)
```

Scaling is a common preprocessing step in machine learning to ensure that the features are on the same scale, which can be important for some machine learning algorithms. For regression problems, it is often desirable to scale or transform both the input and the target variables.

Standardization is a one of the scaling technique wherein it makes the data scale-free by converting the statistical distribution of the data into the below format:
* mean: 0 (zero)
* standard deviation: 1

```python
reg = LinearRegression()     # create linear regression object
reg.fit(x_scaled, y_scaled)  # train the model using the training sets

print(reg.coef_)             # print regression coefficients
print(reg.intercept_, '\n')  # print the intercept 

# print the regression equation
print('f(x) =',
      '%.4f' %reg.intercept_,
      '+ %.4f' %reg.coef_[0,0],
      'x + %.4f' %reg.coef_[0,1],
      'x^2\n')

# print R-squared
print('R2 =', '%.4f' %reg.score(x_scaled, y_scaled))
```

The R2 score helps assess how well the model explains the variance in the target variable. It evaluates the performance of a regression-based model.

In other words, R-squared measures how well the independent variables can predict the outcome variable. It ranges from 0 to 1, with higher values indicating a better fit of the model to the data. An R-squared value of 1 means the model perfectly fits the data and there is no difference between the predicted value and actual value. A value of 0 means there is no relationship.

```python
# a range of house sizes from 800 to 2500 sqft with a step of 50
s1 = np.linspace(min(x), max(x), 50).reshape((-1,1))
s2 = s1 ** 2   # the square of s1

# concatenate w1 and w2 along the second axis
s = np.concatenate((s1,s2),1)

# scale the new data using the same scaler for x
s_scaled = x_scaler.fit_transform(s)

# make predictions on the new data w
y_pred_scaled = reg.predict(s_scaled)

# inverse transform the scaled predictions to get predictions in the original scale
y_pred = y_scaler.inverse_transform(y_pred_scaled)

# plot the data points
plt.scatter(x[:,0], y, marker='x', c='b', label='Actual data')

# plot the predictions
plt.plot(s1, y_pred, c="r", label='Our prediction')

plt.ylabel("Price in 1000s of $") # set the y axis label
plt.xlabel("House size (sqft)")   # set the x axis label
plt.grid()                        # show grid lines
plt.legend()                      # show the legend
plt.show()                        # display the plot
```

It‚Äôs possible to transform the input array in several ways, like using `insert()` from numpy. But the class `PolynomialFeatures` is very convenient for this purpose. `PolynomialFeatures()` generate a new feature matrix consisting of all polynomial combinations of the features with degree less than or equal to the specified degree. For example, if an input sample is two dimensional and of the form [a, b], the degree-2 polynomial features are [1, a, b, a^2, ab, b^2].
* `degree` : int , default=2 - It specifies the maximal degree of the polynomial features.
* `interaction_only` : bool, default=`False` - If `True`, only interaction features are produced: features that are products of at most degree distinct input features, i.e. terms with power of 2 or higher of the same input feature are excluded:
   * included: x[0], x[1], x[0] * x[1], etc.
   * excluded: x[0] ** 2, x[0] ** 2 * x[1], etc.
* `include_bias` : bool, default=`True` - If `True` (default), then include a bias column, the feature in which all polynomial powers are zero (i.e. a column of ones - acts as an intercept term in a linear model).

You can obtain a very similar result when `include_bias` is `True`. By doing that the intercept is already included with the leftmost column of ones, and you don‚Äôt need to include it again when creating the instance of LinearRegression. Thus, you can provide `fit_intercept` = `False` in `LinearRegression()`. You see that now `.intercept_` is zero, but `.coef_` actually contains intercept as its first element. Everything else is the same.

Keep in mind that you need the input to be a two-dimensional array. That‚Äôs why `.reshape()` is used.

```python
x = np.array([1000,1500,1200,1800,2500,800,1600,2400,3100,3800,4200,4500]).reshape((-1,1))
y = np.array([100,230,180,180,170,120,240,195,220,270,260,380])
```

```python
x_scaled = StandardScaler().fit_transform(x.reshape(-1,1))
y_scaled = StandardScaler().fit_transform(y.reshape(-1,1))
```

```python
d = 3  # maximal degree of the polynomial features
poly = PolynomialFeatures(degree=d, include_bias=False)  # generate polynomial and interaction features
poly.fit(x_scaled)                                       # compute number of output features
x3 = poly.transform(x_scaled)                            # transform data to polynomial features (new x array)

# You can also use .fit_transform() to replace the three previous statements with only one
x3 = PolynomialFeatures(degree=d, include_bias=False).fit_transform(x_scaled)
```

```python
model = LinearRegression().fit(x3, y)   # create and fit the model
```

```python
r_sq = model.score(x3,y)
print('R-squared:', r_sq)

b = model.intercept_
w = model.coef_

print('Intercpt:', b)
print('coeficients:', w)
```

If you want to get the predicted response, just use `.predict()`, but remember that the argument should be the modified input x_ instead of the old x. The prediction works almost the same way as in the case of linear regression. It just requires the modified input instead of the original.

```python
y_pred = model.predict(x3)
print('predicted response:\n', y_pred)
```

Use the following code to draw data points and fitted line when there is only one independet varible.

```python
plt.scatter(x,y, marker="x", c="b", label="Actual data")

z = np.linspace(min(x), max(x), 100).reshape((-1,1))    # create some values for the function
z_scaled = StandardScaler().fit_transform(z)

z_ = PolynomialFeatures(degree=d, include_bias=False).fit_transform(z_scaled)  # transform data to polynomial features

zy = model.predict(z_)  # use the model to predict

plt.plot(z, zy, "r", label="Our prediction")

plt.ylabel('Price')
plt.xlabel('Size')
plt.grid()
plt.legend()
plt.show()
```

Now, let's fit higher degrees of polynomial function to the data and plot it.

```python
for d in range(1,7):

    xp = PolynomialFeatures(degree=d, include_bias=False).fit_transform(x_scaled)
    
    model = LinearRegression().fit(xp, y)
    
    plt.scatter(x, y, marker='x', c='b', label='Actual data')

    z = np.linspace(min(x), max(x), 100).reshape((-1,1))    # create some values for the function
    z_scaled = StandardScaler().fit_transform(z)
    z_ = PolynomialFeatures(degree=d, include_bias=False).fit_transform(z_scaled)  # transform data to polynomial features

    zy = model.predict(z_)  # use the model to predict

    plt.plot(z, zy, 'r', label='Our prediction')

    plt.ylabel('Price')
    plt.xlabel('Size')
    plt.title('degree = %d, R2 = %.4f' %(d, model.score(xp, y)))
    plt.grid()
    plt.legend()
    plt.show()
```
