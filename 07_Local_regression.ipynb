## Local Regression

Local regression or local polynomial regression, also known as moving regression, is a generalization of the moving average and polynomial regression. The idea is giving more weights to the closer data.

Local regression, often implemented as locally weighted scatterplot smoothing (LOWESS) or locally weighted regression (LOESS), is a non-parametric method for fitting a smooth curve to a set of data points. It is a type of regression analysis that focuses on the local behavior of the data rather than assuming a global model. It does not have a predefined function to determine the relationship between the dependent and independent variables.

* **Local Weighting:** For each data point in the dataset, a weighted regression is performed, giving more importance to nearby points and less importance to distant points. The weights are determined by a kernel function that assigns higher weights to points closer to the data point of interest.


* **Regression Fit:** A weighted regression is then performed for each data point, considering only the nearby points with higher weights. This is typically done using a polynomial regression model of a specified degree.


* **Smoothing:** The fitted values from the local regressions are combined to create a smooth curve that captures the underlying trend in the data. The degree of smoothing is often controlled by a parameter, which determines the size of the neighborhood considered for each local fit. Adjusting this parameter allows you to control the trade-off between capturing fine details and obtaining a smoother overall trend in the data.

```python
import numpy as np
import math
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# create 100 random sample (x, y)
n = 100
x = np.linspace(0, 2 * math.pi, n)
y = np.sin(x) + 0.3 * np.random.randn(n)
```

```python
# lowess will return our "smoothed" data with a y value for at every x-value
lowess = sm.nonparametric.lowess(y, x, frac=.1)

# unpack the lowess smoothed points to their values
lowess_x = lowess[:,0] # extract x from the lowess result
lowess_y = lowess[:,1] # extract the smoothed y from the lowess result
```

```python
plt.plot(x, y, 'o')  # scatter plot of noisy data
plt.plot(x, lowess_y, 'k', lw=3)  # smoothed curve in black

plt.show()
```

In the context of the `statsmodels.nonparametric.lowess` function, the `frac` parameter represents the fraction of the data used when estimating each y-value in the lowess smoothing process. It determines the size of the local neighborhood for each data point.

* The `frac` parameter takes a value between 0 and 1.
* A smaller `frac` value means that a smaller fraction of the data will be used for each local regression, resulting in more localized and flexible fits.
* A larger `frac` value, on the other hand, means that a larger fraction of the data will be used for each local regression, resulting in smoother but potentially oversimplified fits.

The `frac=0.1` means that the lowess function will use approximately 10% of the data to estimate the y-values at each point along the x-axis.

The following code plot the fitted function with different `frac` values to see how it affects the smoothness of the resulting curve. Adjusting this parameter allows you to control the level of smoothing and the trade-off between capturing local variations and obtaining an overall smooth trend in the data.

```python
plt.figure(figsize=(12, 6))
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.2, hspace=0.4)

plt.subplot(221)
lowess = sm.nonparametric.lowess(y, x, frac=.05)
plt.scatter(x, y, marker='o', label='y true')
plt.plot(x, lowess[:,1], 'k', lw=3, label='y pred')
plt.title('$frac=0.05$')

plt.subplot(222)
lowess = sm.nonparametric.lowess(y, x, frac=.4)
plt.scatter(x, y, marker='o', label='y true')
plt.plot(x, lowess[:,1], 'k', lw=3, label='y pred')
plt.title('$frac=0.40$')

plt.subplot(223)
lowess = sm.nonparametric.lowess(y, x, frac=.5)
plt.scatter(x, y, marker='o', label='y true')
plt.plot(x, lowess[:,1], 'k', lw=3, label='y pred')
plt.title('$frac=0.70$')

plt.subplot(224)
lowess = sm.nonparametric.lowess(y, x, frac=.95)
plt.scatter(x, y, marker='o', label='y true')
plt.plot(x, lowess[:,1], 'k', lw=3, label='y pred')
plt.title('$frac=0.95$')

plt.show()
```
