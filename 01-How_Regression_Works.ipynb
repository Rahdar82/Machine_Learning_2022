# Introduction

Regression is a statistical technique used to analyze the relationship between a dependent variable and one or more independent variables. The goal of regression is to find the best-fitting line or curve that can predict the value of the dependent (target) variable based on the values of the independent variables. Regression can be used for a wide range of applications, such as predicting sales revenue based on advertising spending, estimating the impact of education on income, and modeling the relationship between temperature and crop yields.

Here's a breakdown of how regression analysis works:

- **Variables**: In regression analysis, you have two types of variables:
   - Dependent variable ($y$): This is the variable you want to predict or explain. It's often denoted as $y$.
   - Independent variables ($x$): These are the variables that you believe have an impact on the dependent variable. They're often denoted as $ x_1, x_2, \ldots, x_n $.

- **Simple vs. Multiple Regression**: 
   - Simple Regression: When you have only one independent variable.
   - Multiple Regression: When you have more than one independent variable.

- **Model**: The relationship between the dependent and independent variables is represented by a mathematical equation called the regression model. The basic form of a simple linear regression model is:
   $$ y = b + w x $$
   - $b$ is the intercept (the value of $y$ when $x = 0$)
   - $w$ is the slope of the line (the change in $y$ for a one-unit change in $x$)

The goal is to estimate the coefficients $b$ and $w$ that minimize the difference between the observed values of $y$ and the values predicted by the model ($\hat{y}$). This is typically done using the method of least squares.

Once the model is estimated, it's important to assess its goodness of fit and statistical significance. This involves examining measures like the coefficient of determination ($R^2$), which tells you how well the independent variables explain the variability in the dependent variable, and conducting hypothesis tests on the coefficients to determine if they are significantly different from zero.

To create a regression model:
* Data is collected on both the dependent and independent variables.
* The data is then used to estimate the values of $b$ and $w$ that minimize the sum of the squared errors between the predicted values and the actual values of the dependent variable. This process is known as least squares regression.
* Once the regression model has been estimated, it can be used to predict the value of the dependent variable for a given value of the independent variable.
* The accuracy of the predictions can be evaluated using various statistical measures such as the coefficient of determination (R-squared) and the root-mean-square error (RMSE).

# A Simple Example
The following code snippet represents a simple dataset for training a regression model to predict house prices based on their sizes. It initializes two arrays: `x_train`, containing the sizes of houses in thousands of square feet, and `y_train`, representing the corresponding prices of these houses in thousands of dollars. # Regression Cost Function

Playing with different values of $b$ and $w$ is not an efficient way of finding the best value for these parameters to have a best line fitted to data.

So, we utilize the concept of the cost function. In regression, the cost function quantifies the disparity between predicted and actual values of the dependent variable within the training dataset. The objective is to determine the model parameters that minimize this cost function, ensuring the best fit to the training data.

In linear regression, a commonly employed cost function is the mean squared error (MSE). It is expressed as:

$$ \text{MSE} = \frac{1}{2m} \sum_{i} (y_i - \hat{y}_i)^2 $$

Here, $ m $ represents the number of training examples, $ y_i$ denotes the actual output for the $i^{th}$ training example, and $ \hat{y}_i $ is the predicted output, given by the linear function $ \hat{y}_i = b + w x_i $.

To minimize the cost function and determine the optimal values of $ b $ and $ w $, we can employ an iterative optimization algorithm known as gradient descent. This algorithm adjusts the model parameters iteratively, moving them in the direction of the negative gradient of the cost function. By repetitively updating the parameters, the cost function gradually diminishes until it reaches a minimum, signifying a satisfactory fit of the model to the training data. The specifics of the gradient descent algorithm will be explored in subsequent labs.

```python
b = 2  # intercept
w = 3  # slope

f = lambda x : b + w*x  # fitted line

plt.plot(x, y, 'o', color='k', label='training data')  # data points

plt.plot(x, f(x), c='r', label='y = %.1f + %.1f*x' %(b,w)) # fitted line

for i in range(len(x)):  # gap between actual point and predicted
    plt.plot([x[i], x[i]], [f(x[i]), y[i]], ls='--', color='b', lw=.7)

plt.ylabel('Price in $100k (y)')           # set the y axis label
plt.xlabel('House size in 1k of sqft (x)') # set the x axis label
plt.legend()
plt.show()
```

This visualization helps in assessing how well the linear regression model fits the training data by comparing the actual data points to the fitted line. It also shows the discrepancies between the actual and predicted values for each data point. As you see, our gusse for coefficients of the line are not necessraily work well. Let's assume that $b=0$ and plot the cost for different values of $w$ to see for what $w$ the cost is minimum.

```python
# define a function to calculate cost J
def cost(b, w, x, y):
    """
    Calculates the cost J for a given set of parameters b and w in a linear regression model.
    
    Inputs:
        b: The intercept parameter
        w: The slope parameter
        x: An array containing the independent variable values (house sizes)
        y: An array containing the dependent variable values (prices)
    
    Output:
        J: the average cost for the given parameters b and w
    """
    m = len(x)  # number of data points
    J = 0       # initialize J
    for i in range(m):
        J += ((b + w*x[i]) - y[i] )**2 / m
    return J
```

Let's see how the cost varies with different values of the slope parameter w, while keeping the intercept parameter b fixed at 0 by generating a plot. It can provide insights into the sensitivity of the cost function to changes in the slope parameter, which is useful for understanding the behavior of the regression model and for optimizing the choice of model parameters.

```python
# generate evenly spaced numbers between 0.2 and 5.
w = np.linspace(.2, 5, 20)

# for the fixed b, draw cost for various w between 0.2 and 5
plt.plot(w, cost(0, w ,x ,y))  

plt.xlabel('$w$', size = 14)
plt.ylabel('cost', size = 14)
plt.show()
```

For a simple univariate linear regression model, the cost function depends on changing in two variables: $b$ and $w$. However, for simplicity, we assume that $b=0$, but we should optimize both at the same time. As shown above, if $b=0$, the cost is minimum when $w$ is around 2.5.

```python
fig = plt.figure(figsize=(7,5)) # specify size of 7x5

plt.plot(x,y, 'o', color='k', label='training data')

b = 0
w = np.linspace(0.2, 5 ,5)

for i in range(len(w)):
    plt.plot(x, b + w[i]*x, label='w = %.2f' %w[i] )
    
plt.title('Fitted Lines with Different Slopes', size='large')
plt.ylabel('Price in $100k (y)')           # set the y axis label
plt.xlabel('House size in 1k of sqft (x)') # set the x axis label
plt.legend()
plt.show()
```

Now, let's change b and w at the same time and plot the cost function. This plot provides a visual representation of how the cost function varies with different combinations of intercept and slope parameters, allowing for insights into the optimization landscape of the regression model.

`np.meshgrid` is a NumPy function that returns coordinate matrices from coordinate vectors.
Given two 1D arrays, it returns two 2D matrices. The first matrix represents the X-coordinates of the points on a grid and the second matrix represents the Y-coordinates of the points on a grid.

```python
fig = plt.figure(figsize=(12,8)) # specify size of 12x8

# an object of the Axes3D class, which is used to create a 3D plot
ax = fig.gca(projection='3d')

b = np.linspace(-2,2,100)
w = np.linspace(0,5,100)

B, W = np.meshgrid(b, w)
Z = cost(B, W, x, y)

# create the surface using the B, W, and Z arrays
surf = ax.plot_surface(B, W, Z, cmap=plt.cm.rainbow)

# add a color bar to the plot
fig.colorbar(surf)

# set title and axis labels
ax.set_title('Cost Function Surface')
ax.set_xlabel('Intercept (b)')
ax.set_ylabel('Slope (w)')
ax.set_zlabel('Cost')

plt.show()
```

```python
fig = plt.figure()

# create a contour plot projected onto the z-axis
contour = plt.contour(B, W, Z, levels=[i for i in np.arange(0,80)], cmap=plt.cm.rainbow)

# add a color bar for the contour plot
fig.colorbar(contour)

plt.xlabel('Intercept (b)')
plt.ylabel('Slope (w)')
plt.title('Contour Plot of Cost Function', size='large')
plt.show()
```

Contour plots are useful for visualizing how the value of a function changes over a two-dimensional space. In the context of a cost function in linear regression, a contour plot can provide insights into the optimization landscape by visualizing how the cost varies with different combinations of model parameters (intercept b and slope w). Look for the lowest points (valleys) in the contour plot. These represent combinations of intercept and slope values that minimize the cost function. These optimal parameter values are often the ones you want to use for your regression model.

Observe the contour lines. Closer contour lines indicate regions where the cost changes more rapidly with small changes in parameter values. Conversely, more widely spaced contour lines suggest regions of parameter space where the cost changes more slowly. This can help you understand the sensitivity of the cost function to changes in model parameters.
