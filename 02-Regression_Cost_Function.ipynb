# Regression cost function

In regression, the cost function (also known as the loss function) is a mathematical expression that measures the difference between the predicted values of the dependent variable and the actual values in the training dataset. The goal of regression is to find the values of the model parameters that minimize the cost function and provide the best fit to the training data.

The cost function for a regression model typically measures the difference between the predicted output of the model $f(x)$ and the actual output. A commonly used cost function for linear regression is the mean squared error (MSE), which is defined as: 
$MSE = \frac{1}{2m} \sum_i (y_i - f(x_i))^2$

where $m$ is the number of training examples, $y_i$ is the actual output for the i-th training example, and $f(x_i)$ is the predicted output for the i-th training example ($f(x_i) = b+wx_i$). The factor of $\frac{1}{2m}$ is included to simplify the calculation of the gradient during the optimization process.

The cost function is minimized by adjusting the model's parameters (e.g., weights $w$ and biases $b$) during training to minimize the difference between the predicted and actual output.

To minimize the cost function and find the optimal values of $b$ and $w$, gradient descent can be used. Gradient descent is an iterative algorithm that updates the model parameters in the direction of the negative gradient of the cost function. So the cost function is gradually reduced until it reaches a minimum and the model provides a good fit to the training data. We will learn about gradient descent algorithm in future labs.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
sns.set()   # setting the default Seaborn theme
```

Lets try an example. We generate random values for $y$ when $x$ is integers from 1 to 10. Then, we try different values for coeficients of line $f(x) = b + wx$ to estimate the cost of our prediction.

```python
# some simple random (close to linear) data
x = np.arange(1,11)
y = x + 1.5 * np.random.randn(10)
```

```python
plt.plot(x, y, 'o', color='k', label='training data');  # data points

plt.xlabel('x', size = 14)
plt.ylabel('y', size = 14)
plt.show()
```

```python
b = 0     # intercept
w = 0.5   # slope
f = lambda x : b + w*x  # fitted line

plt.plot(x, y, 'o', color='k', label='training data')  # data points

plt.plot(x, f(x), label='f(x) = %.1f + %.1f*x' %(b,w)) # fitted line

for i in range(len(x)):  # gap between actual point and predicted
    plt.plot([x[i], x[i]], [f(x[i]), y[i]], ls='--', color='r', lw=.7)

plt.xlabel('x', size = 14)
plt.ylabel('y', size = 14)
plt.legend()
plt.show()
```

As you see, our gusse for coefficients of the line are not necessraily work well. Let's assume that $b=0$ and plot the cost for different values of $w$ to see for what $w$ the cost is minimum.

```python
# a function to calculate cost J
def cost(b, w, x, y):
    m = len(x)  # number of data points
    J = 0       # initialize J
    for i in range(m):
        J += ((b + w*x[i]) - y[i] )**2
    return J/m
```

```python
# generate evenly spaced numbers between 0.2 and 2.
w = np.linspace(.2, 2, 20)

# for the fixed b, draw cost for various w between 0.2 and 2
plt.plot(w, cost(0, w ,x ,y))  

plt.xlabel('$w$', size = 14)
plt.ylabel('cost', size = 14)
plt.show()
```

For a simple univariate linear regression model, the cost function depends on changing in two variables: $b$ and $w$. However, for simplicity, we assume that $b=0$, but we should optimize both at the same time. As shown above, if $b=0$, the cost is minimum when $w$ is around 1. 

```python
fig = plt.figure(figsize=(7,5)) # specify size of 7x5

plt.plot(x,y, 'o', color='k', label='training data')

w = np.linspace(0.2, 2 ,5)

for i in range(len(w)):
    plt.plot(x, 0+w[i]*x, label='w = %.2f' %w[i] )
    
plt.legend()
plt.show()
```

`np.meshgrid` is a NumPy function that returns coordinate matrices from coordinate vectors.
Given two 1D arrays, it returns two 2D matrices. The first matrix represents the X-coordinates of the points on a grid and the second matrix represents the Y-coordinates of the points on a grid.

```python
fig = plt.figure(figsize=(12,8)) # specify size of 12x8

# an object of the Axes3D class, which is used to create a 3D plot
ax = fig.gca(projection='3d')

b = np.linspace(-2,2,100)
w = np.linspace(0,2.2,100)

B, W = np.meshgrid(b, w)
Z = cost(B, W, x, y)

# create the surface using the B, W, and Z arrays
surf = ax.plot_surface(B, W, Z, cmap=plt.cm.rainbow)
fig.colorbar(surf) # # add a color bar to the plot

plt.show()
```

```python
fig = plt.figure()

# create a contour plot projected onto the z-axis
plt.contour(B, W, Z, levels=[i for i in np.arange(0,80)], cmap=plt.cm.rainbow)
fig.colorbar(surf)

plt.xlabel('b', size = 14)
plt.ylabel('w', size = 14)
plt.show()
```

As you see from the above contour plot, we can see what combinations of $b$ and $w$ gives the least cost. We will learn how to find the optimal value for $b$ and $w$ in the next lab.
